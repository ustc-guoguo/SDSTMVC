```python
from torch.nn import functional as F
import torch
emb1 = torch.Tensor([1,2,3,2,1])
emb2 = torch.Tensor([1,2,3,2,1])
m = (F.softmax(emb1, dim=-1) + F.softmax(emb2, dim=-1)) / 2

kl_p_m = F.kl_div(F.log_softmax(emb1, dim=-1), m, reduction='batchmean')
kl_q_m = F.kl_div(F.log_softmax(emb2, dim=-1), m, reduction='batchmean')

js_loss = 0.5 * (kl_p_m + kl_q_m)

print(js_loss)
```
